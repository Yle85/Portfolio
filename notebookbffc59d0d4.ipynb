```python
# Cocoa percent and choco bars

plt.figure(figsize=(15,5))
choco_data['Cocoa\nPercent'].value_counts().head(10).sort_index().plot.bar(color=['#d9d9d8','#b3b3b4','#808080','#000000','#404040','#d9d9d8','#b3b3b3','#404040','#b3b3b4'])
plt.xlabel('Percentage of Cocoa',size=12,color='brown')
plt.ylabel('Number of Chocolate bars',size=12,color='brown')
plt.show()
```

```python
x= range(50)
y = range(50) + np.random.randint(0,30,50)
plt.scatter(x, y, c=y, cmap='Spectral')
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('Rating by Cocoa Percent')
plt.xlabel('Cocoa\nPercent')
plt.ylabel('Rating')
plt.show()
```
